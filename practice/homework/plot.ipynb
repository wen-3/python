```python
import math
def uni_hist(N):
    x = np.random.uniform(0.0, 1.0, N)
    y = np.random.uniform(0.0, 1.0, N)

# 將圖形分割為多個
    plt.subplot(2,1,1)
    plt.hist(x)
    plt.title("No1:histogram")
    
    plt.subplot(2,1,2)
    plt.hist(y)
    plt.title("No2:histogram")    
    
    plt.grid(True)
    
    plt.tight_layout

# N = 1000
uni_hist(1000)
```
